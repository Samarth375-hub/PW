```python
def calculate_results(marks):
    
    print("Marks in five major subjects:")
    for i, mark in enumerate(marks, start=1):
        print(f"Subject {i}: {mark}")
    
    # Calculate the sum of marks
    total_marks = sum(marks)
    print(f"Total Marks: {total_marks}")
    
    # Calculate the percentage
    percentage = total_marks / 5
    print(f"Percentage: {percentage}%")
    
    # Determine the grade using match-case
    match percentage:
        case p if p >85:
            grade = 'A'
        case p if 85>p >=75:
            grade = 'B'
        case p if 75 > p >=50:
            grade = 'C'
        case p if 30 < p <= 50:
            grade = 'D'
        case _:
            grade = 'Reappear'
        
    
    # Display the grade
    print(f"Grade: {grade}")

# Example usage
marks = [85, 78, 92, 88, 76]  # Replace with actual marks
calculate_results(marks)
```

```python
def find_color_by_wavelength(wavelength):
    match wavelength:
        case wl if 400 <= wl <= 440:
            color = "Violet"
        case wl if 440 < wl <= 460:
            color = "Indigo"
        case wl if 460 < wl <= 500:
            color = "Blue"
        case wl if 500 < wl <= 570:
            color = "Green"
        case wl if 570 < wl <= 590:
            color = "Yellow"
        case wl if 590 < wl <= 620:
            color = "Orange"
        case wl if 620 < wl <= 720:
            color = "Red"
        case _:
            color = "Wavelength out of visible range"
    
    return color

# Example usage
wavelengths = [400, 460, 500, 570, 590, 600, 720]

print("VIBGYOR Spectrum Colors:")
for wl in wavelengths:
    color = find_color_by_wavelength(wl)
    print(f"Wavelength {wl} nm: {color}")
```

```python
def gravitational_force(m1, m2, r):
    G = 6.674 * (10 ** -11) 
    force = G * (m1 * m2) / (r ** 2)
    return force

def compare_forces():
    # Masses in kilograms
    mass_earth = 5.972 * (10 ** 24)  # Mass of Earth
    mass_moon = 7.347 * (10 ** 22)   # Mass of Moon
    mass_sun = 1.989 * (10 ** 30)    # Mass of Sun

    # Distances in meters
    distance_earth_moon = 3.844 * (10 ** 8)  # Distance between Earth and Moon
    distance_earth_sun = 1.496 * (10 ** 11)  # Distance between Earth and Sun

    # Calculate gravitational forces
    force_earth_moon = gravitational_force(mass_earth, mass_moon, distance_earth_moon)
    force_earth_sun = gravitational_force(mass_earth, mass_sun, distance_earth_sun)

    # Print the forces
    print(f"Gravitational Force between Earth and Moon: {force_earth_moon:.2e} N")
    print(f"Gravitational Force between Earth and Sun: {force_earth_sun:.2e} N")

    # Compare the forces
    if force_earth_sun > force_earth_moon:
        stronger_force = "Sun"
    else:
        stronger_force = "Moon"

    print(f"The gravitational force between Earth and {stronger_force} is stronger.")

# Run the comparison
compare_forces()
```

```python
class LibraryBook:
    def __init__(self, book_name, author, is_available=True):
        # Encapsulated attributes
        self.__book_name = book_name
        self.__author = author
        self.__is_available = is_available
    
    # Getter for book name
    def get_book_name(self):
        return self.__book_name
    
    # Getter for author
    def get_author(self):
        return self.__author
    
    # Getter for availability status
    def is_book_available(self):
        return self.__is_available
    
    # Method to borrow a book
    def borrow_book(self):
        if self.__is_available:
            self.__is_available = False
            print(f"The book '{self.__book_name}' by {self.__author} has been borrowed.")
        else:
            print(f"Sorry, the book '{self.__book_name}' by {self.__author} is currently unavailable.")
    
    # Method to return a book
    def return_book(self):
        if not self.__is_available:
            self.__is_available = True
            print(f"The book '{self.__book_name}' by {self.__author} has been returned.")
        else:
            print(f"The book '{self.__book_name}' by {self.__author} was not borrowed.")

    # Method to display book details
    def display_info(self):
        status = "Available" if self.__is_available else "Unavailable"
        print(f"Book: {self.__book_name}, Author: {self.__author}, Status: {status}")

# Example usage
book1 = LibraryBook("To Kill a Mockingbird", "Harper Lee")
book2 = LibraryBook("1984", "George Orwell", False)

# Display book information
book1.display_info()
book2.display_info()

# Borrowing books
book1.borrow_book()
book2.borrow_book()

# Returning books
book1.return_book()
book2.return_book()

# Display updated book information
book1.display_info()
book2.display_info()
```

```python
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0):
        self.__account_number = account_number
        self.__account_holder = account_holder
        self.__balance = balance
    
    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self.__balance:.2f}")
        else:
            print("Deposit amount must be positive.")
    
    # Method to withdraw money
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew ${amount:.2f}. New balance: ${self.__balance:.2f}")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")
    
    # Method to check balance
    def check_balance(self):
        print(f"Account balance: ${self.__balance:.2f}")
    
    # Method to display account details
    def display_account_info(self):
        print(f"Account Number: {self.__account_number}")
        print(f"Account Holder: {self.__account_holder}")
        self.check_balance()

class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0, interest_rate=0.02):
        super().__init__(account_number, account_holder, balance)
        self.__interest_rate = interest_rate
    
    # Method to apply interest
    def apply_interest(self):
        interest = self._BankAccount__balance * self.__interest_rate
        self.deposit(interest)
        print(f"Applied interest: ${interest:.2f}")

class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0, overdraft_limit=500):
        super().__init__(account_number, account_holder, balance)
        self.__overdraft_limit = overdraft_limit
    
    # Method to withdraw money with overdraft limit
    def withdraw(self, amount):
        if amount > 0:
            if amount <= (self._BankAccount__balance + self.__overdraft_limit):
                self._BankAccount__balance -= amount
                print(f"Withdrew ${amount:.2f}. New balance: ${self._BankAccount__balance:.2f}")
            else:
                print("Overdraft limit exceeded.")
        else:
            print("Withdrawal amount must be positive.")

# Example usage
def main():
    # Create instances of SavingsAccount and CheckingAccount
    savings = SavingsAccount("12345", "Alice", 1000, 0.03)
    checking = CheckingAccount("67890", "Bob", 500, 200)
    
    # Display account information
    print("Savings Account Info:")
    savings.display_account_info()
    
    print("\nChecking Account Info:")
    checking.display_account_info()
    
    # Perform transactions
    print("\nDepositing $200 to Savings Account...")
    savings.deposit(200)
    
    print("\nWithdrawing $100 from Checking Account...")
    checking.withdraw(100)
    
    print("\nApplying interest to Savings Account...")
    savings.apply_interest()
    
    print("\nWithdrawing $750 from Checking Account...")
    checking.withdraw(750)
    
    print("\nFinal Savings Account Info:")
    savings.display_account_info()
    
    print("\nFinal Checking Account Info:")
    checking.display_account_info()

# Run the main function
if __name__ == "__main__":
    main()
```

```python
class Room:
    _id_counter = 1  # Class-level counter to generate unique room IDs

    def __init__(self, room_number, room_type, rate):
        self.__id = Room._id_counter  # Unique room ID
        Room._id_counter += 1
        self.__room_number = room_number
        self.__room_type = room_type
        self.__rate = rate
        self.__availability = True

    # Getter for unique room ID (private)
    def get_id(self):
        return self.__id

    # Getter and Setter for room number
    def get_room_number(self):
        return self.__room_number

    def set_room_number(self, room_number):
        self.__room_number = room_number

    # Getter and Setter for room type
    def get_room_type(self):
        return self.__room_type

    def set_room_type(self, room_type):
        self.__room_type = room_type

    # Getter and Setter for rate
    def get_rate(self):
        return self.__rate

    def set_rate(self, rate):
        self.__rate = rate

    # Check availability
    def is_available(self):
        return self.__availability

    # Book room
    def book_room(self):
        if self.__availability:
            self.__availability = False
            print(f"Room {self.__room_number} booked successfully.")
        else:
            print(f"Room {self.__room_number} is already booked.")

    # Check-in guest
    def check_in(self):
        if self.__availability:
            print(f"Room {self.__room_number} is available for check-in.")
        else:
            print(f"Room {self.__room_number} is not available for check-in.")

    # Check-out guest
    def check_out(self):
        if not self.__availability:
            self.__availability = True
            print(f"Room {self.__room_number} checked out successfully.")
        else:
            print(f"Room {self.__room_number} is already available.")

class SuiteRoom(Room):
    def __init__(self, room_number, rate, num_bedrooms):
        super().__init__(room_number, "Suite", rate)
        self.__num_bedrooms = num_bedrooms

    # Getter and Setter for number of bedrooms
    def get_num_bedrooms(self):
        return self.__num_bedrooms

    def set_num_bedrooms(self, num_bedrooms):
        self.__num_bedrooms = num_bedrooms

    # Display room info including number of bedrooms
    def display_info(self):
        print(f"Room Number: {self.get_room_number()}")
        print(f"Room Type: {self.get_room_type()}")
        print(f"Rate: ${self.get_rate():.2f}")
        print(f"Number of Bedrooms: {self.__num_bedrooms}")
        print(f"Availability: {'Available' if self.is_available() else 'Booked'}")

class StandardRoom(Room):
    def __init__(self, room_number, rate, has_tv):
        super().__init__(room_number, "Standard", rate)
        self.__has_tv = has_tv

    # Getter and Setter for TV availability
    def has_tv(self):
        return self.__has_tv

    def set_tv(self, has_tv):
        self.__has_tv = has_tv

    # Display room info including TV availability
    def display_info(self):
        print(f"Room Number: {self.get_room_number()}")
        print(f"Room Type: {self.get_room_type()}")
        print(f"Rate: ${self.get_rate():.2f}")
        print(f"TV Available: {'Yes' if self.__has_tv else 'No'}")
        print(f"Availability: {'Available' if self.is_available() else 'Booked'}")

# Example usage
def main():
    # Create room instances
    suite = SuiteRoom("101", 150.00, 2)
    standard = StandardRoom("102", 80.00, True)

    # Display room information
    print("Suite Room Info:")
    suite.display_info()
    
    print("\nStandard Room Info:")
    standard.display_info()

    # Book and check-in rooms
    print("\nBooking and Checking In Suite Room...")
    suite.book_room()
    suite.check_in()

    # Attempt to check-in again (should indicate it's not available)
    print("\nAttempting to Check-In Suite Room Again...")
    suite.check_in()

    # Check-out suite room
    print("\nChecking Out Suite Room...")
    suite.check_out()

    # Display updated information
    print("\nUpdated Suite Room Info:")
    suite.display_info()

# Run the main function
if __name__ == "__main__":
    main()
```

```python
class Member:
    _id_counter = 1  # Class-level counter to generate unique member IDs

    def __init__(self, name, age, membership_type):
        self.__id = Member._id_counter  # Unique member ID
        Member._id_counter += 1
        self.__name = name
        self.__age = age
        self.__membership_type = membership_type
        self.__membership_status = 'Active'  # Default membership status

    # Getter for unique member ID (private)
    def get_id(self):
        return self.__id

    # Getter and Setter for name
    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    # Getter and Setter for age
    def get_age(self):
        return self.__age

    def set_age(self, age):
        self.__age = age

    # Getter and Setter for membership type
    def get_membership_type(self):
        return self.__membership_type

    def set_membership_type(self, membership_type):
        self.__membership_type = membership_type

    # Getter and Setter for membership status
    def get_membership_status(self):
        return self.__membership_status

    def set_membership_status(self, status):
        self.__membership_status = status

    # Register a new member
    def register_member(self):
        print(f"Member {self.__name} (ID: {self.__id}) registered with {self.__membership_type} membership.")

    # Renew membership
    def renew_membership(self):
        if self.__membership_status == 'Active':
            print(f"Member {self.__name} (ID: {self.__id}) already has an active membership.")
        else:
            self.__membership_status = 'Active'
            print(f"Member {self.__name} (ID: {self.__id}) membership renewed.")

    # Cancel membership
    def cancel_membership(self):
        if self.__membership_status == 'Cancelled':
            print(f"Member {self.__name} (ID: {self.__id}) membership is already cancelled.")
        else:
            self.__membership_status = 'Cancelled'
            print(f"Member {self.__name} (ID: {self.__id}) membership cancelled.")

class FamilyMember(Member):
    def __init__(self, name, age, membership_type, family_size):
        super().__init__(name, age, membership_type)
        self.__family_size = family_size

    # Getter and Setter for family size
    def get_family_size(self):
        return self.__family_size

    def set_family_size(self, family_size):
        self.__family_size = family_size

    # Display family member info
    def display_info(self):
        print(f"Family Member ID: {self.get_id()}")
        print(f"Name: {self.get_name()}")
        print(f"Age: {self.get_age()}")
        print(f"Membership Type: {self.get_membership_type()}")
        print(f"Membership Status: {self.get_membership_status()}")
        print(f"Family Size: {self.__family_size}")

class IndividualMember(Member):
    def __init__(self, name, age, membership_type, occupation):
        super().__init__(name, age, membership_type)
        self.__occupation = occupation

    # Getter and Setter for occupation
    def get_occupation(self):
        return self.__occupation

    def set_occupation(self, occupation):
        self.__occupation = occupation

    # Display individual member info
    def display_info(self):
        print(f"Individual Member ID: {self.get_id()}")
        print(f"Name: {self.get_name()}")
        print(f"Age: {self.get_age()}")
        print(f"Membership Type: {self.get_membership_type()}")
        print(f"Membership Status: {self.get_membership_status()}")
        print(f"Occupation: {self.__occupation}")

# Example usage
def main():
    # Create member instances
    family_member = FamilyMember("John Doe", 40, "Family", 4)
    individual_member = IndividualMember("Jane Smith", 30, "Individual", "Engineer")

    # Register and display member info
    family_member.register_member()
    family_member.display_info()
    
    individual_member.register_member()
    individual_member.display_info()

    # Renew and cancel membership
    print("\nRenewing Family Member's Membership...")
    family_member.renew_membership()
    print("\nCancelling Individual Member's Membership...")
    individual_member.cancel_membership()

    # Display updated member info
    print("\nUpdated Family Member Info:")
    family_member.display_info()
    
    print("\nUpdated Individual Member Info:")
    individual_member.display_info()

# Run the main function
if __name__ == "__main__":
    main()
```

```python
class Event:
    _id_counter = 1  # Class-level counter to generate unique event IDs

    def __init__(self, name, date, time, location):
        self.__id = Event._id_counter  # Unique event ID
        Event._id_counter += 1
        self.__name = name
        self.__date = date
        self.__time = time
        self.__location = location
        self.__attendees = []

    # Getter for unique event ID (private)
    def get_id(self):
        return self.__id

    # Getter and Setter for name
    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    # Getter and Setter for date
    def get_date(self):
        return self.__date

    def set_date(self, date):
        self.__date = date

    # Getter and Setter for time
    def get_time(self):
        return self.__time

    def set_time(self, time):
        self.__time = time

    # Getter and Setter for location
    def get_location(self):
        return self.__location

    def set_location(self, location):
        self.__location = location

    # Add attendee
    def add_attendee(self, attendee):
        if attendee not in self.__attendees:
            self.__attendees.append(attendee)
            print(f"Attendee {attendee} added to event '{self.__name}'.")
        else:
            print(f"Attendee {attendee} is already on the list.")

    # Remove attendee
    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"Attendee {attendee} removed from event '{self.__name}'.")
        else:
            print(f"Attendee {attendee} is not in the list.")

    # Get total number of attendees
    def get_total_attendees(self):
        return len(self.__attendees)

class PrivateEvent(Event):
    def __init__(self, name, date, time, location, invited_guests):
        super().__init__(name, date, time, location)
        self.__invited_guests = invited_guests

    # Getter and Setter for invited guests
    def get_invited_guests(self):
        return self.__invited_guests

    def set_invited_guests(self, invited_guests):
        self.__invited_guests = invited_guests

    # Display event info
    def display_info(self):
        print(f"Event ID: {self.get_id()}")
        print(f"Name: {self.get_name()}")
        print(f"Date: {self.get_date()}")
        print(f"Time: {self.get_time()}")
        print(f"Location: {self.get_location()}")
        print(f"Invited Guests: {self.__invited_guests}")
        print(f"Total Attendees: {self.get_total_attendees()}")

class PublicEvent(Event):
    def __init__(self, name, date, time, location, registration_fee):
        super().__init__(name, date, time, location)
        self.__registration_fee = registration_fee

    # Getter and Setter for registration fee
    def get_registration_fee(self):
        return self.__registration_fee

    def set_registration_fee(self, registration_fee):
        self.__registration_fee = registration_fee

    # Display event info
    def display_info(self):
        print(f"Event ID: {self.get_id()}")
        print(f"Name: {self.get_name()}")
        print(f"Date: {self.get_date()}")
        print(f"Time: {self.get_time()}")
        print(f"Location: {self.get_location()}")
        print(f"Registration Fee: ${self.__registration_fee:.2f}")
        print(f"Total Attendees: {self.get_total_attendees()}")

# Example usage
def main():
    # Create event instances
    private_event = PrivateEvent("Exclusive Gala", "2024-09-15", "19:00", "Grand Hotel", ["Alice", "Bob"])
    public_event = PublicEvent("Community Fair", "2024-10-01", "10:00", "City Park", 25.00)

    # Add and remove attendees
    print("Managing Private Event...")
    private_event.add_attendee("Charlie")
    private_event.remove_attendee("Alice")
    private_event.display_info()
    
    print("\nManaging Public Event...")
    public_event.add_attendee("Dave")
    public_event.add_attendee("Eve")
    public_event.remove_attendee("Dave")
    public_event.display_info()

# Run the main function
if __name__ == "__main__":
    main()
```

```python
class Flight:
    _id_counter = 1  # Class-level counter to generate unique flight IDs

    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats):
        self.__id = Flight._id_counter  # Unique flight ID
        Flight._id_counter += 1
        self.__flight_number = flight_number
        self.__departure_airport = departure_airport
        self.__arrival_airport = arrival_airport
        self.__departure_time = departure_time
        self.__arrival_time = arrival_time
        self.__total_seats = total_seats
        self.__available_seats = total_seats

    # Getter for unique flight ID (private)
    def get_id(self):
        return self.__id

    # Getter and Setter for flight number
    def get_flight_number(self):
        return self.__flight_number

    def set_flight_number(self, flight_number):
        self.__flight_number = flight_number

    # Getter and Setter for departure airport
    def get_departure_airport(self):
        return self.__departure_airport

    def set_departure_airport(self, departure_airport):
        self.__departure_airport = departure_airport

    # Getter and Setter for arrival airport
    def get_arrival_airport(self):
        return self.__arrival_airport

    def set_arrival_airport(self, arrival_airport):
        self.__arrival_airport = arrival_airport

    # Getter and Setter for departure time
    def get_departure_time(self):
        return self.__departure_time

    def set_departure_time(self, departure_time):
        self.__departure_time = departure_time

    # Getter and Setter for arrival time
    def get_arrival_time(self):
        return self.__arrival_time

    def set_arrival_time(self, arrival_time):
        self.__arrival_time = arrival_time

    # Getter for available seats
    def get_available_seats(self):
        return self.__available_seats

    # Book a seat
    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked successfully. Remaining seats: {self.__available_seats}.")
        else:
            print("No available seats.")

    # Cancel a reservation
    def cancel_reservation(self):
        if self.__available_seats < self.__total_seats:
            self.__available_seats += 1
            print(f"Reservation cancelled. Remaining seats: {self.__available_seats}.")
        else:
            print("No reservations to cancel.")

class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, state_code):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.__state_code = state_code

    # Getter and Setter for state code
    def get_state_code(self):
        return self.__state_code

    def set_state_code(self, state_code):
        self.__state_code = state_code

    # Display flight info
    def display_info(self):
        print(f"Flight ID: {self.get_id()}")
        print(f"Flight Number: {self.get_flight_number()}")
        print(f"Departure Airport: {self.get_departure_airport()}")
        print(f"Arrival Airport: {self.get_arrival_airport()}")
        print(f"Departure Time: {self.get_departure_time()}")
        print(f"Arrival Time: {self.get_arrival_time()}")
        print(f"State Code: {self.__state_code}")
        print(f"Available Seats: {self.get_available_seats()}")

class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, country_code):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.__country_code = country_code

    # Getter and Setter for country code
    def get_country_code(self):
        return self.__country_code

    def set_country_code(self, country_code):
        self.__country_code = country_code

    # Display flight info
    def display_info(self):
        print(f"Flight ID: {self.get_id()}")
        print(f"Flight Number: {self.get_flight_number()}")
        print(f"Departure Airport: {self.get_departure_airport()}")
        print(f"Arrival Airport: {self.get_arrival_airport()}")
        print(f"Departure Time: {self.get_departure_time()}")
        print(f"Arrival Time: {self.get_arrival_time()}")
        print(f"Country Code: {self.__country_code}")
        print(f"Available Seats: {self.get_available_seats()}")

# Example usage
def main():
    # Create flight instances
    domestic_flight = DomesticFlight("DL123", "LAX", "SFO", "09:00", "10:30", 150, "CA")
    international_flight = InternationalFlight("AI789", "JFK", "LHR", "18:00", "06:00", 200, "GB")

    # Book and cancel reservations
    print("Managing Domestic Flight...")
    domestic_flight.book_seat()
    domestic_flight.cancel_reservation()
    domestic_flight.display_info()
    
    print("\nManaging International Flight...")
    international_flight.book_seat()
    international_flight.cancel_reservation()
    international_flight.display_info()

# Run the main function
if __name__ == "__main__":
    main()
```

```python
# Question 14

def reverse_string(s):
    """Reverses the given string."""
    return s[::-1]

def capitalize_string(s):
    """Capitalizes the first letter of each word in the string."""
    return s.title()

# Example usage
if __name__ == "__main__":
    s = "hello world"
    print("Original String:", s)
    print("Reversed String:", reverse_string(s))
    print("Capitalized String:", capitalize_string(s))
```

```python
# file_operations.py
# question 15
def write_to_file(filename, data):
    """Writes the given data to the file."""
    with open(filename, 'w') as file:
        file.write(data)

def read_from_file(filename):
    """Reads the content of the file."""
    with open(filename, 'r') as file:
        return file.read()

def append_to_file(filename, data):
    """Appends the given data to the file."""
    with open(filename, 'a') as file:
        file.write(data)

# Example usage
if __name__ == "__main__":
    filename = "example.txt"
    data = "Hello, World!\n"
    
    write_to_file(filename, data)
    print("Data written to file.")
    
    append_to_file(filename, "This is an appended line.\n")
    print("Data appended to file.")
    
    content = read_from_file(filename)
    print("File Content:\n", content)
```

```python
# question 16
# employees.py

def write_employee_details(filename, employees):
    """Writes employee details to the file."""
    with open(filename, 'w') as file:
        for emp in employees:
            file.write(f"Name: {emp['name']}, Age: {emp['age']}, Salary: {emp['salary']}\n")

# Example usage
if __name__ == "__main__":
    employees = [
        {"name": "John Doe", "age": 30, "salary": 50000},
        {"name": "Jane Smith", "age": 25, "salary": 60000},
        {"name": "Sam Johnson", "age": 35, "salary": 55000}
    ]
    
    write_employee_details("employees.txt", employees)
    print("Employee details written to file.")
```

```python
# question 18.
# expenses_calculator.py

def calculate_total_expenses(filename):
    """Calculates the total amount spent based on the expenses listed in the file."""
    total = 0
    with open(filename, 'r') as file:
        for line in file:
            amount = float(line.strip())
            total += amount
    return total

# Example usage
if __name__ == "__main__":
    total = calculate_total_expenses("expenses.txt")
    print(f"Total Amount Spent: {total}")
```

```python
# Question 19.
# word_counter.py

from collections import Counter

def count_words(filename):
    """Counts the occurrences of each word in the file and displays them in alphabetical order."""
    with open(filename, 'r') as file:
        text = file.read().lower()
    
    words = text.split()
    word_count = Counter(words)
    
    for word in sorted(word_count.keys()):
        print(f"{word}: {word_count[word]}")

# Example usage
if __name__ == "__main__":
    count_words("paragraph.txt")
```
